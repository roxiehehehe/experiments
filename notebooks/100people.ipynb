There are out 100 people in a room and they each have 100 dollars. Every minute each person will randomly give another person 1 dollar. What is the wealth distribution after x minutes?

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib

from matplotlib import animation, rc
from IPython.display import HTML

%matplotlib inline 
# matplotlib.rcParams['figure.figsize'] = (14, 7)
NUM_PEOPLE = 10
```

```python
# Init
people = [NUM_PEOPLE]*NUM_PEOPLE
fig, ax = plt.subplots()


ax.set_xlim(( -1, NUM_PEOPLE+1))
ax.set_ylim((0, NUM_PEOPLE * 3))
ax.set_xlabel('People ID')
ax.set_ylabel('Wealth')
x = np.linspace(0, NUM_PEOPLE, NUM_PEOPLE)


bars = ax.bar(x, people)
    
```

```python
# initialization function: plot the background of each frame
def init():
    for bar in bars:
        bar.set_height(NUM_PEOPLE)
    return bars
```

```python
def exchange(i,j):
    # i will give j $1 if i has money
    if people[i] > 0:
        people[i] -= 1
        people[j] += 1
```

```python
# animation function. This is called sequentially
def animate(i):
    
    i = random.randrange(0, NUM_PEOPLE)
    j = random.randrange(0, NUM_PEOPLE)
    exchange(i,j)
    
    for bar, h in zip(bars, sorted(people)):
        bar.set_height(h)
        
    return bars
```

```python
# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=500, interval=10, blit=True, repeat=False)
```

```python
HTML(anim.to_html5_video())
```
